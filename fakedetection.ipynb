```python
from sklearn.model_selection import train_test_split
```

```python
feature_names=fdata.iloc[:,1:30].columns
target=fdata.iloc[:1,30:].columns
print(feature_names)
print(target)
```

```python
x=fdata[feature_names]
y=fdata[target]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
print("the length of the xtrian:{xtrain}".format(xtrain=len(x_train)))
print("the length of the xtest:{xtest}".format(xtest=len(x_test)))
print("the length of the ytrian:{ytrain}".format(ytrain=len(y_train)))
print("the length of the xtrian:{ytest}".format(ytest=len(y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
lmodel=LogisticRegression()
lmodel.fit(x_train,y_train)
```

```python
pred=lmodel.predict(x_test)
```

```python
pred
```

```python
class_name=["not_fraud","fraud"]
matrix=confusion_matrix(y_test,pred)
dataframe=pd.DataFrame(matrix,index=class_name,columns=class_name)
sns.heatmap(dataframe,annot=True,cbar=None,cmap="Blues",fmt="g")
plt.title("confusion matrix"),
plt.tight_layout(),
plt.ylabel("True class")
plt.xlabel("Predicted class")
plt.show()
```

```python
from sklearn.metrics import f1_score,recall_score
recall_score=round(recall_score(y_test,pred),2)
f1_score=round(f1_score(y_test,pred),2)
print("Sensitivity/Recall for Logistic Regression Model 1 : {recall_score}".format(recall_score = recall_score))
print("Sensitivity/Recall for Logistic Regression Model 1 : {f1_score}".format(f1_score = f1_score))
      
      

```
